````python
import pandas as pd
# use encoding while importing the csv file
df = pd.read_csv("/content/drive/MyDrive/Diwali Sales Data.csv",encoding = 'unicode_escape')
# first 10 rows it will show
df.head(10)
````

````python
# since the last two columns are null so we are going to remove them
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
# inplace = ture (means store the resultant dataset after modification)
# axis = 1 (delete that row)
````

````python
df.info()
````

````python
pd.isnull(df)
````

````python
# to check if any null value
pd.isnull(df).sum()
````

````python
# remove null values
df.dropna(inplace=True)
df.info()
pd.isnull(df).sum()
````

````python
df['Amount'] = df['Amount'].astype('int')
# changing data type of amount column from float to int
df['Amount'].dtypes
````

````python
# if we want to rename any column
df.rename(columns = {'Marital_Status':'IsMarried'})
````

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


# using count plot
# 1st parameter is x axis
# 2nd parameter (data) = dataFrame_Name

ax = sns.countplot(x = 'Gender',data=df)
````

````python
ax = sns.countplot(x = 'Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
````

````python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
````

````python
ds = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=ds)
````

from the above graphs we can see that most buyers are females

````python
# orders by each state
sales_set = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(15)
# .head(15) to top 15 states only
sns.set(rc={'figure.figsize':(25,5)})
ax1 = sns.barplot(data=sales_set,y='Orders',x='State')

for bars in  ax1.containers:
  ax1.bar_label(bars)
````
